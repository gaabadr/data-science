```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, palette='coolwarm')
```

```python
#tip = gorjeta
#total_bill = valor da conta


#gráfico: há uma relação linear, uma correlação entre o total da conta e o valor da gorjeta
#quanto maior o valor da conta, maior tenderá ser o valor da gorjeta

sns.lmplot(x='total_bill',y='tip', data=tips, palette='coolwarm', hue='sex', scatter_kws={'s':50})
```

```python
#as maiores gorjetas são dadas por homens

```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, palette='coolwarm',row='time',hue='sex',scatter_kws={'s':70})
```

```python
#existe uma probabilidade de a quantidade de gorjetas e os seus valores serem maiores no período do jantar

sns.boxplot(x='day',y='tip',data=tips, hue='sex')
```

```python
#as maiores gorjetas são dadas no sábado
#a maior QUANTIDADE de gorjetas altas são dadas no domingo
#obs: os pontos dispersos acima do gráfico no sábado e na terça são OUTLIER: um valor atípico, fora da média
```

```python
#podemos fazer uma análise mais completa com a função abaixo:

sns.pairplot(tips)
```

```python
#concluímos que a única relação linear que existe é entre o valor da conta e o da gorjeta
```
