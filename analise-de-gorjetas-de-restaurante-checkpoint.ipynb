```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, palette='coolwarm')
```

```python
#tip = gorjeta
#total_bill = valor da conta


#grfico: h uma relao linear, uma correlao entre o total da conta e o valor da gorjeta
#quanto maior o valor da conta, maior tender ser o valor da gorjeta

sns.lmplot(x='total_bill',y='tip', data=tips, palette='coolwarm', hue='sex', scatter_kws={'s':50})
```

```python
#as maiores gorjetas so dadas por homens

```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, palette='coolwarm',row='time',hue='sex',scatter_kws={'s':70})
```

```python
#existe uma probabilidade de a quantidade de gorjetas e os seus valores serem maiores no perodo do jantar

sns.boxplot(x='day',y='tip',data=tips, hue='sex')
```

```python
#as maiores gorjetas so dadas no sbado
#a maior QUANTIDADE de gorjetas altas so dadas no domingo
#obs: os pontos dispersos acima do grfico no sbado e na tera so OUTLIER: um valor atpico, fora da mdia
```

```python
#podemos fazer uma anlise mais completa com a funo abaixo:

sns.pairplot(tips)
```
