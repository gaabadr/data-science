```python
from pandas_datareader import data,wb
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)

BAC = data.DataReader('BAC','yahoo', start,end)

C = data.DataReader('C','yahoo', start,end)
GS = data.DataReader('GS','yahoo', start,end)
JPM = data.DataReader('JPM','yahoo', start,end)
MS = data.DataReader('MS','yahoo', start,end)
WFC = data.DataReader('WFC','yahoo', start,end)
```

```python
tickers=['BAC','C','GS','JPM','MS','WFC']


bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC], axis=1,keys=tickers)
```

```python
bank_stocks.columns.names=['Bank Ticker','Stock Info']
bank_stocks.head()
```

```python
#qual o preço máximo de fechamento para o estoque de cada banco durante o período?

bank_stocks.xs(key='Close',axis=1, level='Stock Info').head()
```

```python
bank_stocks.xs(key='Close',axis=1, level='Stock Info').max() #retorna o valor máximo do fechamento de cada banco
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick + 'Return'] = bank_stocks[tick]['Close'].pct_change()# pct_change calcula a alteração percentual
returns.head()                                                                      #serve para comparar a porcentagem de alteração em uma série temporal de elementos
```

```python
#vamos analisar um desses retornos:
returns['CReturn'].plot()
```

```python
# a ação mais arriscada para todo o período de tempo é a do banco C

#qual a ação mais arriscada para o ano 2015?

returns[ (returns.index.date>=datetime.date(2015,1,1)) & (returns.index.date<datetime.date(2016,1,1)) ].std()
```

```python
# a mais arriscada para 2015 é a do banco MS
# vamos criar um gráfico de distribuição desse banco para esse período

sns.distplot(returns[ (returns.index.date>=datetime.date(2015,1,1)) & (returns.index.date<datetime.date(2016,1,1)) ]['MSReturn'])

```
